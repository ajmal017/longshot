```python
import pandas as pd
from tqdm import tqdm
from database.strategy import Strategy
from database.market import Market
from transformer.column_transformer import ColumnTransformer
from transformer.date_transformer import DateTransformer
import warnings
warnings.simplefilter(action='ignore', category=Warning)
import matplotlib.pyplot as plt 
from datetime import datetime, timedelta, timezone
import math
import numpy as np
```

```python
strat_db = Strategy("unity")
market = Market()
```

```python
strat_db.connect()
predictions = strat_db.retrieve_data("quarterly_fundamental_sim")
strat_db.close()
```

```python
for ticker in predictions["ticker"].unique():
    market.connect()
    prices = market.retrieve_price_data("pdr_prices",ticker)
    prices["year"] = [x.year for x in prices["Date"]]
    prices["quarter"] = [x.quarter for x in prices["Date"]]
    prices = prices.groupby(["year","quarter"]).mean()
    market.close()
    ticker_data = predictions[predictions["ticker"]==ticker]
    
    vs = ticker_data.merge(prices,on=["year","quarter"],how="left")
    vs["pdiff"] = vs["quarterly_fundamental_regression_prediction"].pct_change()
    vs["adiff"] = vs["Adj Close"].pct_change()
    plt.plot(vs['quarterly_fundamental_regression_prediction'])
    plt.plot(vs['Adj Close'])
    plt.title(ticker)
    plt.show()
```
