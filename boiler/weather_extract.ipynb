```python
import pandas as pd 
from datetime import datetime, timedelta
from extractor.weather_extractor import WeatherExtractor
from database.weather import Weather
```

```python
cities = ["New York"
          ,"Los Angeles"
          ,"Chicago"
          ,"San Francisco"
          ,"Washington D.C"
          ,"Dallas"
          ,"Houston"
          ,"Boston"
          ,"Philadelphia"
          ,"Atlanta"
          ,"Seattle"
          ,"Miami"] 
```

```python
start = "2009-01-01"
end ="2020-11-01"
weather_db = Weather()
```

```python
weather_db.connect()
weather_db.drop_table("weather")
weather_db.close()
timeline = pd.date_range(start,end)
years = timeline.year
months = timeline.month
dates = pd.DataFrame({"timeline":timeline,"year":years,"month":months})
for year in dates["year"].unique():
    year_dates = dates[dates["year"] == year]
    for month in year_dates["month"].unique():
        month_dates = year_dates[year_dates["month"] == month]
        start_date = str(month_dates.head(1)["timeline"].item())
        end_date = str(month_dates.tail(1)["timeline"].item())
        print(start_date,end_date)
        for city in cities:
            try:
                wex = WeatherExtractor(city,start_date,end_date)
                data = wex.extract()
                weather = pd.DataFrame(data["data"]["weather"])
                weather["precipMM"] = [row[1]["hourly"][0]["precipMM"] for row in weather.iterrows()]
                rc = ["date"]
                for col in weather.columns:
                    if "precip" in col or "max" in col or "min" in col:
                        rc.append(col) 
                final = weather[rc]
                final["location"] = city
                weather_db.connect()
                weather_db.store_data("weather",final)
                weather_db.close()
            except Exception as e:
                print(city,str(e))
```
